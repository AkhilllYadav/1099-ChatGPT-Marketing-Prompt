## **Image Data**

```python
import matplotlib.pyplot as plt
import imageio
from skimage import io, color
```

```python
colored_image = imageio.imread('/content/sample_data/Pale blue dot.jpg')

gray_image = color.rgb2gray(colored_image)
print(gray_image)
```

```python
plt.figure(figsize=(6,6))
plt.imshow( gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()

print(gray_image.shape)
```

```python
plt.figure(figsize=(6,6))
plt.imshow(colored_image)
plt.title('Colored Image')
plt.axis('off')
plt.show()
print(colored_image.shape)
```

```python
import cv2

image1 = imageio.imread('/content/sample_data/Pale blue dot.jpg')
image2 = imageio.imread('/content/sample_data/209015877.jpg')

print(image1.shape)
print(image2.shape)
```

```python
if image1 is None or image2 is None:
    print("Error: Unable to read one or both images.")
else:
    image2_resized = cv2.resize(image2, (image1.shape[1],image1.shape[0]))
    print(image2_resized.shape)
```

```python
plt.imshow(image1)
plt.show()

plt.imshow(image2)
plt.show()
```

```python
added_image = image1 + image2_resized

plt.imshow(added_image)
plt.show()
```

```python
subtracted_image12 = image1 - image2_resized

plt.imshow(subtracted_image12)
plt.show()
```

```python
subtracted_image21 = image2_resized - image1

plt.imshow(subtracted_image21)
plt.show()
```
